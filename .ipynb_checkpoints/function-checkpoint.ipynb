# Feature Selection

```python
df.drop(['District_Name','Crop_Year'],axis=1,inplace=True)
```

```python
df.rename(columns={'Season':'Period'},inplace=True)
```

```python
df['Crop'].nunique()
```

```python
df['Period'].value_counts()
```

```python
df_final=pd.get_dummies(df)
```

```python
df_final.head()
```

```python
df_final.info()
```

```python
df_final.columns=df_final.columns.str.r
```

## model creation

```python
x=df_final.drop('Production',axis=1)
y=df_final['Production']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=5)
```

```python
x.info()
```

```python
linear_model=LinearRegression()
linear_model.fit(x_train,y_train)
```

```python
#model evaluation on traning data 
y_train_pred=linear_model.predict(x_train)
y_train_pred[0:5]
```

```python
y_train[0:5]
```

```python
# checking the accuracy
linear_accuracy=r2_score(y_train,y_train_pred)
linear_accuracy
```

```python
# model evaluation on testing data 
y_test_pred=linear_model.predict(x_test)
y_test_pred[0:5]
```

```python
y_test[0:5]
```

```python
# checking the accuracy
linear_accuracy_test=r2_score(y_test,y_test_pred)
linear_accuracy_test
```

# Decision Tress

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(random_state=42)
regressor.fit(x_train,y_train)
```

```python
#model evaluation on traning data 
y_train_pred_dt=regressor.predict(x_train)
y_train_pred_dt[0:5]
```

```python
y_train[0:5]
```

```python
# checking the accuracy
regressor_accuracy_train_dt=r2_score(y_train,y_train_pred_dt)
regressor_accuracy_train_dt
```

```python
# model evaluation on testing data 
y_test_pred_dt=regressor.predict(x_test)
y_test_pred_dt[0:5]
```

```python
y_test[0:5]
```

```python
# checking the accuracy
regressor_accuracy_test_dt=r2_score(y_test,y_test_pred_dt)
regressor_accuracy_test_dt
```

## Random Forest

```python
rf_model=RandomForestRegressor(random_state=35)
rf_model.fit(x_train,y_train)
```

```python
# model evaluation for traning data 
y_train_pred_rf=rf_model.predict(x_train)
y_train_pred_rf[0:5]
```

```python
# accuracy on traning data 
rf_accuracy_train=r2_score(y_train,y_train_pred_rf)
rf_accuracy_train
```

```python
# model Evalution on test data 
y_test_pred_rf=rf_model.predict(x_test)
y_test_pred_rf[0:5]
```

```python
y_test[0:5]
```

```python
# accuracy on testing data 
rf_accuracy_test=r2_score(y_test,y_test_pred_rf)
rf_accuracy_test
```

```python
# hyper parameter tunning for Random Forest 
model_rf_hyp = RandomForestRegressor(random_state=3)

hyp = {
    'n_estimators':np.arange(50,200,10),
    'criterion':['squared_error','absolute_error'],
    'max_depth':np.arange(5,10),
    'min_samples_split':np.arange(2,10),
    'min_samples_leaf':np.arange(2,10)
}

rscv = RandomizedSearchCV(model_rf_hyp,hyp, cv = 6)
rscv.fit(x_train,y_train)
rscv.best_estimator_
```

```python
# accuracy on testing data 
ada_accuracy_test=r2_score(y_test,y_test_pred_ada)
ada_accuracy_test
```

## Hyperprarameters tunning for adaboost

```python
model_adaboost_hyp = AdaBoostRegressor()

hyp = {
    "n_estimators": np.arange(10,150,10),
    "learning_rate":np.arange(0.1,2,0.1),
    "random_state" :np.arange(1,10)
}

rscv = RandomizedSearchCV(model_adaboost_hyp,hyp, cv = 5)
rscv.fit(x_train,y_train)
```
